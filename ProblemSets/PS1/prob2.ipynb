```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f1(n):
    return np.log(n)

def f2(n):
    return n ** np.sqrt(n)

def f3(n):
    return np.log(n) ** np.log(n)

def f4(n):
    return n ** (3/4)

def f5(n):
    return np.log(np.log(n))

def f6(n):
    return 2 ** np.log2(n)

def f7(n):
    return np.full_like(n, 10)

def f8(n):
    return 5 ** n

def f9(n):
    return np.log2(n) * n

def f10(n):
    return (np.log2(n)) ** n

# 生成 x 值
x = np.linspace(1, 500, 100)

# 计算各函数值
y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
y4 = f4(x)
y5 = f5(x)
y6 = f6(x)
y7 = f7(x)
y8 = f8(x)
y9 = f9(x)
y10 = f10(x)

# 绘图
plt.figure(figsize=(10, 8))
plt.plot(x, y1, label='$f_1(n) = \log_2 n$')
plt.plot(x, y2, label='$f_2(n) = n^{sqrt{n}}$')
plt.plot(x, y3, label='$f_3(n) = \log(n)^{\log(n)}$')
plt.plot(x, y4, label='$f_4(n) = n^{3/4}$')
plt.plot(x, y5, label='$f_5(n) = \log_2(\log_2 n)$')
plt.plot(x, y6, label='$f_6(n) = 2^{\log_2 n}$')
plt.plot(x, y7, label='$f_7(n) = 10$')
plt.plot(x, y8, label='$f_8(n) = 5^n$')
plt.plot(x, y9, label='$f_9(n) = \log_2 n \cdot n$')
plt.plot(x, y10, label='$f_{10}(n) = (\log_2 n)^n$')
plt.yscale('log')  # 使用对数坐标轴
plt.xlabel('n')
plt.ylabel('f(n)')
plt.ylim([0, 10**12])
plt.title('Comparing Growth Rates of Functions')
plt.legend()
plt.grid(True)
plt.show()
```
